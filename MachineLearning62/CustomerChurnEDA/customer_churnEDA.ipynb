Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the data

```python
df = pd.read_csv("Customer_churn_dataset.csv")
```

```python
df.head()
```

Total Number of customers

```python
df["Customer_ID"].count()
```

Number of customers by gender

```python
df.groupby("Gender")["Customer_ID"].count()
```

Average Spending Score of customers

```python
df["Spending_Score"].mean()
```

What is most frequent spending score

```python
df["Spending_Score"].median()
```

Average Spending Score by gender

```python
df.groupby("Gender")["Spending_Score"].mean()
```

Highest 10 Customers by purchase count

```python
df.sort_values(by="Purchase_Count",ascending=False).head(10)[["Customer_ID","Purchase_Count"]]
```

Top 10 customer with highest annual income

```python
df[["Customer_ID","Annual_Income"]].sort_values(by="Annual_Income",ascending=False).head(10)
```

Top 10 customer with highest income and their gender should be female, churn must be True

```python
Female_data = df[(df["Gender"]=="Female") & (df["Churn"]==1)][["Customer_ID","Annual_Income"]]
```

```python
Female_data.sort_values(by="Annual_Income",ascending=False).head(10)
```
