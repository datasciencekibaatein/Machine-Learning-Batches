<h2>What is time series?</h2>

A time series is data collected in sequence over time intervals. Where time order is exteremely important.

<h2>Components of time series</h2>

1. Trend
2. Seasonality
3. Cycles
4. Noise

<h2>1. Trend</h2>
Long term increase or decrease

<h2>2. Seasonality</h2>

Pattern repeated at fixed intervals.

<h2>3. Cyclic Pattern</h2>

Rise/fall over irregular long periods.

<h2>4. Noise</h2>
Random unpredictable fluctuations.

<h2>Frequency in time series</h2>

D -> Daily<br>
W -> Weekly<br>
M -> Monthly<br>
Q -> Quarterly<br>
Y -> Yearl<br>
H -> hourly<br>

```python
import pandas as pd

dates = pd.date_range(start='2025-01-01',periods=6,freq='M')

sales = [100,120,130,150,170,180]

df = pd.DataFrame({
    'Dates':dates,
    'Sales':sales
})

df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Dates'],df['Sales'])
plt.show()
```

```python
df['Rolling_mean'] = df['Sales'].rolling(3).mean()
```

```python
df['Lag1'] = df['Sales'].shift(1)
df['Lag2'] = df['Sales'].shift(2)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['Sales'],order=(1,1,1))

fit = model.fit()

forecast = fit.forecast(5)
print(forecast)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(df['Sales'],order=(1,1,1),seasonal_order=(1,1,1,12))

fit = model.fit()

forecast = fit.forecast(5)
print(forecast)
```
