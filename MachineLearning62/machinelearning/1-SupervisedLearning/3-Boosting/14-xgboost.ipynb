<h1><strong>XgBoost</strong></h1>

XGBoost (Extreme gradient boosting) is a powerful machine learning algorithm based on gradient boosting used for:

- Classification
- Regression
- Ranking

XGBoost solved these by introducting:

- Regularizatoin
- Parallel tree construction
- Handling missing values
- Faster optimization

XGBoost = Extreme Gradient Boosting

<strong>How XGBoost Works:</strong>

- Step 01: Inital Prediction
- Step 02: Compute the residuals (residual = actual points - predicted points)
- Step 03: Train tree on Residual
- Step 04: Update the prediction
- Step 05: Repeat 

<strong>Important Hyperparameters</strong>

- n_estimators (number of trees)
- learning_rate (controls contribution of each tree) Small value = slower learning but better performace
- max_depth (maximum tree depth)
- subsample (percentage of rows used)
- colsample_bytree (precentage of columns used)
- gamma (minimum loss reducntion required to split)
- lambda (L2 Regularization)
- alpha (L1 Regularizatino)

<h3><strong>Advantages of XGBoost</strong></h3>

- very high accuracy
- handles missing values
- prevents overfitting
- parallel processing
- works well on tabular data

<h3><strong>Disadvantages</strong></h3>

- Harder to interpret
- Many hyperparameters
- can overfit small datasets

```python
import xgboost as xgb
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = xgb.XGBClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

<strong>Hyper parameter tune</strong>

```python
model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8
)

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

<h3><code>Regression Problem</code></h3>

```python
import xgboost as xgb
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = fetch_california_housing()
X= data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = xgb.XGBRegressor()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, pred)
```
