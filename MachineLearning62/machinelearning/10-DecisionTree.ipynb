<center><h1><strong>Decision Tree Classifier</strong></h1></center>

Decision tree algorithm is a supervised machine learning algorithm used for:
- Classification
- Regression

Types of Decision Tree

- Classification Tree(yes/no, 1/0)
- Regression Tree(predicts continous values eg price, marks, salary)

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
data = load_breast_cancer()

x = data.data
y = data.target

print("Fearure shape: ",x.shape)
print("Target shape: ",y.shape)
```

train test split the data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

train the decision tree model

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
```

Prediction

```python
y_pred = model.predict(x_test)
```

Evaluete model

```python
print("Accuracy: ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plot_tree(model,filled=True)
plt.show()
```
