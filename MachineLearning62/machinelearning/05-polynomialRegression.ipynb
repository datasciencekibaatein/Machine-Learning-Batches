<center><h1><strong>Polynomial Regression</strong></h1></center>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 100)

noise = np.random.normal(0, 15, size=len(x))
y = 2*x**3 - 3*x**2 + 5*x + 10 + noise
```

```python
plt.scatter(x,y,color="red")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

<h3>Apply linear regression</h3>

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x.reshape(100,1),y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

Create regression model

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
model = reg.fit(x_train,y_train)
```

Predict the x test

```python
y_pred = model.predict(x_test)
```

Now check the r2 score

```python
from sklearn.metrics import r2_score

score = r2_score(y_test,y_pred)
```

```python
score
```

plot the data poinst and line

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color="red")
plt.xlabel("x test data points")
plt.ylabel("y test data points")
plt.show()
```

<h2><strong>Apply Polynomial regression</strong></h2>

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3,include_bias=True)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
poly_model = regression.fit(x_train_poly,y_train)
```

```python
y_pred_poly = poly_model.predict(x_test_poly)
```

```python
y_pred_poly
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred_poly)
```

```python
score
```

Plot the polynomial

```python
plt.scatter(x_train,y_train)
plt.scatter(x_test,y_pred_poly,color="red")
plt.xlabel("x data points")
plt.ylabel("y data points")
plt.show()
```
