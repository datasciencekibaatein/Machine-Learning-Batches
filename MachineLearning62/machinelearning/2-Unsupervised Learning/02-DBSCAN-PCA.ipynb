<h1><strong>DBSCAN</strong></h1>

DBSCAN groups data based on density instead of distance.

- Points close together -> same cluster
- Sparse regions -> noise(outliers)

|K-means | DBSCAN |
|------|----|
|Need number of clusters (k)| No need to specify clusters|
|Fail with irregular shapes|Detects arbitrary shapes|
|Sensitive to outliers | handles outliers automatically|

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,_ = make_moons(n_samples=300,noise=0.05)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=5)
labels = dbscan.fit_predict(X)
```

min_samples:

- Small -> More clusters
- Large -> Stricter density -> Fewer clusters

eps:

- Small -> more clusters/noise
- Large -> Fewer clusters

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.title("DBSCAN Clustering")
plt.show()
```

<h3><strong>how to choose the eps?</strong></h3>

```python
from sklearn.neighbors import NearestNeighbors

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances,indices =  neighbors_fit.kneighbors(X)

distances = np.sort(distances[:,4])
plt.plot(distances)
plt.title("K distances")
plt.show()
```

<h1><strong>PCA(Principal component analysis)</strong></h1>

PCA is a dimensionality reduction technique

It transforms:
- high dimensional data -> Lower Dimensional
- Whihle keeping maximum variance

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
X
```

```python
x_pca
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.title("PCA Projection (2D)")
plt.xlabel("PC1")
plt.xlabel("PC2")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.09,min_samples=5)
labels = dbscan.fit_predict(x_pca)
plt.scatter(x_pca[:,0],x_pca[:,1],c=labels)
plt.show()
```
