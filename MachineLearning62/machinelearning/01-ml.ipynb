<h2><strong>What is machine learning ?</strong><h2>

<p>Machine learning is a subfield of AI that enables machines to learn from data without being explicitly programmed.</p>

<strong>Examples</strong>
- Fraud/spam detection
- Recommendation systems
- Filtering email

<h3><strong>Types of Machine learning</strong><h3>

1. Supervised learning
2. Unsupervised learning
3. Reinforcement Learning
4. Semi- supervised learning

<strong>Supervised Learning</strong>

Supervised learning uses labeled dataset to train algorithms to understand data patterns and predict outcomes. 
For example: Spam/Fraud detection

The supervised learning further can be classified into two types- Classifcation and Regression

- Linear Regression
- Logistic Regression
- Decision Tree
- Random Forest
- K-nearest Neigbor
- Support Vector Machine
- Naive Bayes

<strong>Unsupervised Learning</strong>
Unsupervised learning is a type of machine learning that uses unlabeled dataset to discover patterns without any explicit guidance or instruction. 

For Example: Customer Segmentation

- K-means Clustering
- PCA
- Hierarchical Clustering
- DBSCAN

<strong>Reinforcement learning</strong>
Reinforcement learning is type of ML where an model learns by interacting with an environment and receiving rewards.

For Example: Training robot

- Q-learning
- SARSA
- DQN

<strong>Semi-supervised</strong>

Semi supervised learning (SSL) it is combination of supervised and unsupervised ML.

Lableled Data = few points
Unlableled Data = many points

For example: Medical Images
