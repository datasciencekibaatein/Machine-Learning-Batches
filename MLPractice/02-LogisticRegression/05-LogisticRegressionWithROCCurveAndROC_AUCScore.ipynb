```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
X,y = make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=2,random_state=15)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

Create dummy model with default 0 as output

```python
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
#Prediction based on probability
model_prob = model.predict_proba(x_test)
```

```python
#focust on positive outcome
model_prob = model_prob[:,1]
```

```python
dummy_model_auc = roc_auc_score(y_test,dummy_model_prob)
model_auc = roc_auc_score(y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
#calculate ROC 
dummy_fpr, dummy_tpr,_ = roc_curve(y_test,dummy_model_prob)
model_fpr, model_tpr,_ = roc_curve(y_test,model_prob)
```

```python
plt.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy model')
plt.plot(model_fpr,model_tpr,marker=".",label='logistic')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

- orange line representing logistic regression
- this area below the orange line , maximum amount of area is catured compared to dummy model
- area below the blue line if for dummy model

```python
#calculate ROC 
dummy_fpr, dummy_tpr,_ = roc_curve(y_test,dummy_model_prob)
model_fpr, model_tpr,threshold = roc_curve(y_test,model_prob)
```

```python

fig, ax = plt.subplots(figsize=(20,50))  # ✅ create fig + ax together

ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy model')
ax.plot(model_fpr, model_tpr, marker=".", label='Logistic')

# annotate thresholds
for xyz in zip(model_fpr, model_tpr, threshold):
    ax.annotate(str(np.round(xyz[2], 2)), xy=(xyz[0], xyz[1]))

ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")
ax.legend()

plt.show()
```
