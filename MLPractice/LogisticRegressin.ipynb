<h2><strong>What is Logistic Regression?</strong></h2>

Logistic Regression is a supervised machine learning algorithm used for classification problems.

Unlike linear regression which predicts continuous values, logistic regression predicts probabilities of classes.

Example problems:

|Problem	|Target|
|-------|------|
|Email spam detection	|Spam / Not Spam|
|Disease prediction	|Sick / Healthy|
|Customer churn	|Leave / Stay|
|Fraud detection	|Fraud / Not Fraud|

So logistic regression answers:

What is the probability that this data belongs to class 1?

<h2><strong>Why Do We Need Logistic Regression?</strong></h2>

If we try to use Linear Regression for classification, problems occur.

Example:

Suppose we predict if a student passes (1) or fails (0).

Linear regression might predict:

- 1.2

- -0.3

But probabilities must lie between:

0≤P≤1

Linear regression cannot guarantee that.

Therefore we use a function that compresses values between 0 and 1.

This is where the Sigmoid Function comes in.

<h2><strong>Sigmoid Function (Logistic Function)</strong></h2>

The sigmoid function converts any real value into a probability.

![sigmoid diagram](/Machine-Learning-Batches/MLPractice/diagrams/img01.JPG)

P(y=1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x)}}

Where:

|Symbol|	Meaning|
|------|-------|
|β0	|Intercept|
|β1	|Coefficient|
|x	|input feature|

The sigmoid curve looks like an S-shaped curve.

Properties:
- Output between 0 and 1
- Used to represent probability

Decision boundary:

- if probability >= 0.5 → Class 1
- if probability < 0.5 → Class 0
