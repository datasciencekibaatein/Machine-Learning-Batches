```python
df
```

<h3>Numerical Value Imputation</h3>

```python
from sklearn.impute import SimpleImputer
```

<h4>handling with mean</h4>

```python
imputer = SimpleImputer(strategy='mean')

df[['Age_filled'] ]= imputer.fit_transform(df[['Age']])
```

```python
df
```

</h4>Handling with median</h4>

```python
median_imputer = SimpleImputer(strategy='median')

df[['Attendance_filled']] = median_imputer.fit_transform(df[['Attendance']])
```

```python
df
```

<h4>Handling with Constant value </h4>

```python
df['Age_constant']= df['Age'].fillna(20)
```

```python
df
```

<h4>End of distribution</h4>

```python
val = df['Age'].mean() + 3 * df['Age'].std()

df['Age_endDist'] = df['Age'].fillna(val)
```

```python
df
```

<h1>Multivariate Imputation</h1>

<h3>KNN - K Nearest Neighbors</h3>

```python
import numpy as np
import pandas as pd

from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    'Experience':[1,2,3,10,11,12,20,21,np.nan],
    'Salary':[45000,50000,np.nan,110000,120000,125000,200000,np.nan,210000],
    'Age':[22,24,23,35,np.nan,38,50,52,51]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
scaler = StandardScaler()

df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled,columns=df.columns)
df_scaled
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
df_imputed = imputer.fit_transform(df_scaled)
```

```python
df_inverse = scaler.inverse_transform(df_imputed)
df_final = pd.DataFrame(df_inverse,columns=df.columns)
df_final
```

<h3>Iterative Imputer</h3>

```python
from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)

age = np.random.randint(20,60,100)
income = (age*20) + np.random.normal(0,5000,100)

df = pd.DataFrame({'Age':age,'Income':income})
```

```python
df.head()
```

```python
df.iloc[1:10,0] = np.nan
df.iloc[20:30,1] = np.nan
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df),columns= df.columns)
```

```python
df_scaled.head(20)
```

```python
it_imputer = IterativeImputer(max_iter=10)

df_imputed = it_imputer.fit_transform(df_scaled)

pd.DataFrame(scaler.inverse_transform(df_imputed),columns=df.columns).head()
```
