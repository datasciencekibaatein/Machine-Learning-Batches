```python
import pandas as pd
```

```python
rng = default_rng(42)

n = 100

age = rng.integers(18,60,size=n)
salary = rng.integers(20000,120000,size=n)
gender = rng.choice(['male','female','other'],size=n)
purchased_prob = (age/60) + (salary/120000)
print(purchased_prob[:6])
purchase_prob = purchased_prob / purchased_prob.max()
print(purchase_prob[:6])
purchased = rng.binomial(1,purchase_prob)

df = pd.DataFrame({
    "Age":age,
    "Salary":salary,
    "Gender":gender,
    "Purchased":purchased
})
```
