<h1><strong>Inferential Statistics</strong></h1>

Inferential statistics helps us to make conclustions about population using a sample.

<strong>Population:</strong>All Observations, All customers of Amazon.<br>
<strong>Sample:</strong>5000 selected customers.


Instead of analyzing whole population, we analyze a sample and infer patterns.

Inferential statistics is the branch of statistics used to:

- estimate populaiton parameters
- test hypotheses
- Make prediction
- Draw conclusion from sample data

<h2>Central Limit Theorm(CLT)</h2>

If sample size is large enough n>=30 then sample means become normally distributed.

```python
import numpy as np

import matplotlib.pyplot as plt

population = np.random.exponential(scale=2,size=100000)

sample_means = []

for _ in range(1000):
    sample = np.random.choice(population,50)
    sample_means.append(sample.mean())


plt.hist(sample_means,bins=30)
plt.title("Samling distribution of means")
plt.show()
```

<h2><strong>Estimation</strong></h2>

1. Point Estimation
2. Interval Estimation

<h3><strong>Point Estimation</strong></h3>

using a single value estimate

Sample mean estimates population mean.



```python
import numpy as np

sample = np.array([10,12,14,16,18])

estimate = np.mean(sample)

print(estimate)
```

<h3><strong>Interval Estimation</strong></h3>

Instead of one value , it gives a range.

We are 95% confident that population mean lies between 50 and 60.

```python
import scipy.stats as sts

import numpy as np
np.random.seed(42)
sample = np.random.normal(50,10,100)

mean = np.mean(sample)
std = np.std(sample,ddof=1)


ci = sts.t.interval(
    0.95,
    len(sample)-1,
    loc=mean,
    scale=sts.sem(sample)
)

print(ci)
```
