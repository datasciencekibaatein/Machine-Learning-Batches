<h2><strong>Measure of central tendency</strong></h2>

Every dataset has a "center of gravity" a single number that represents where the bulk of values live.

<h3><strong>Mean</strong></h3>
The sum of all values divided by the count of values. We call it average

<span style="color:green;">When to use:</span> when data is roughly symmetric and free of exteme values.

<h3><strong>Median</strong></h3>
Sort the values smallest to largest. The median is the middel one . if there are even number of values, take the average of the two middle values.

<span style="color:green;">When to use:</span> When data is skewed or has outliers.

<h3><strong>Mode</strong></h3>
The value that appears most often in the dataset. A dataset may have one mode, two modes, or none at all.

<span style="color:green;">When to use:</span> When you have categorical data


<h2><strong>Measure of dispersion</strong></h2>

Two classes can have the exact same average marks (central tendency) 70 , but one class might be uniformly average (everyone got 68 to 72) while other might be widely mixed (some got 30 , some got 100). The mean alone hides this situation.

 <h3><strong>Range: </strong></h3>

 The simplest spread measure. <span style="color:magenta">Range = Maximum value - Minimum value</span>

 <span style="color:green">Limitation: </span> One outlier completely dominates the range.


 <h3><strong>Variance: </strong></h3>

 The average of the squared differences from the mean. It tells us how far, on average, each value sits from the centre.

 <span style="color:green">Limitation: </span> Because we squared the differences, the unit of variance is the unit of tha data squared (eg, dollor-squared) which is hard to interpred


  <h3><strong>Standard Devation: </strong></h3>
  The square root of variance. This brings the unit back to the original, rupees,marks, kilogram - so it is easy to read.

   <span style="color:green">Limitation: </span> It returns you the approximate equal length of the difference

<h3><strong>Skewness</strong></h3>
Skewness tell us whether the data leans more to one side that other.
    
    - Symmetric (Bell curve)
    - Positive (Right skewed)(long tail on the right)
    - Negative (Left skewed) (long tail on the left)


<h3><strong>Kurtosis</strong></h3>
 It describes often extreme values appear compared to a normal belll curve.
    
    - High kurtosis(data has many extreme outliers)
    - Low kurtosis(data has hardly any extreme values.)

```python
import pandas as pd
import numpy as np

np.random.seed(42)
data = {
    'House':['h1','h2','h3','h4','h5','h6','h7','h8','h9','h10'],
    'Price':[45,50,52,48,55,60,58,49,51,350]
}

df = pd.DataFrame(data)
```

```python
df
```

<h2>Measure of central tendency</h2>

```python
mean_price = df['Price'].mean()
median_price = df['Price'].median()
mode_price = df['Price'].mode()[0]

print(f"Mean: {mean_price}")
print(f"Median: {median_price}")
print(f"Mode: {mode_price}")
```

<h2>Measure of dispersion</h2>

```python
range_val = df['Price'].max() - df['Price'].min()
variance = df['Price'].var()
std_dev = df['Price'].std()
q1,q3 = df['Price'].quantile([0.25,0.75])
iqr = q3 - q1

print(f"Range: {range_val}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Quarter 1 25percentile:  {q1}")
print(f"Quarter 3 75percentile:  {q3}")
print(f"IQR:  {iqr}")

```

<h2>Shape of distribution</h2>

```python
skewness = df['Price'].skew()
kurt = df['Price'].kurtosis()

print(f"Skewness: {skewness}")
print(f"kurtosis: {kurt}")
```
