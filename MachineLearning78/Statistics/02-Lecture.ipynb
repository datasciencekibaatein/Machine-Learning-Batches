<h2><strong>Conditional probability</strong</h2>

<code>Probability</code> measures the chance that an event will occur.

Probability lies  between 0(impossible event) - 1(certain event)

- Probability of getting head in a fair coin toss = 0.5
- Probability of rolling a 6 on a dice = 1/6

<code>Conditional probability</code> The probability of an event happening GIVEN that another event has already happend.

- Does patient has disease ? , The test is positive
- What is the probabiltity an email is spam given certain words appear?

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

<h2><strong>Intution</strong></h2>

- Total Students = 100
- Students who play cricket = 40
- Students who play cricket and know Python = 30

If a student already plays cricket, what is the probability they know python?

P(Python | Cricket) = 30/40 = 0.75, Among cricket players, 75% know python.

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = pd.DataFrame({
    'Purchased':np.random.choice(['Yes','No'],100),
    'Clicked Ad':np.random.choice(['Yes','No'],100)
})

data.head()
```

```python
clicked_yes = data[data['Clicked Ad']=='Yes']

Purchsed_and_clicked_yes = clicked_yes[clicked_yes['Purchased']=='Yes']

Purchsed_and_clicked_yes.shape
```

```python

conditional_probability = len(Purchsed_and_clicked_yes)/len(clicked_yes)
```

```python
conditional_probability
```

```python
A = np.array([1,1,0,1,0,1])
B = np.array([1,0,1,1,0,1])


intersection = np.sum((A==1) & (B==1))

condition = np.sum(B==1)
probability = intersection / condition
probability
```

```python
import matplotlib.pyplot as plt

categories = ['Clicked','Not Clicked']
probablities = [0.7,0.3]

plt.bar(categories,probablities)

plt.ylabel("Probability")
plt.title("Conditional probability")
plt.show()
```
