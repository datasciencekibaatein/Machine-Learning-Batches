```python
temp = pd.merge(pd.merge(sales,customers,on='customer_key'),products,on='product_key')
```

```python
df = temp[['order_number',
           'order_date',
           'shipping_date',
           'sales_amount',
           'quantity',
           'product_number',
           'product_name',
           'category',
           'maintenance',
           'cost',
           'product_line',
           'start_date',
           'customer_number',
           'first_name',
           'last_name',
           "country",
           'marital_status',
           'gender',
           'create_date']]
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['shipping_date'] = pd.to_datetime(df['shipping_date'])
df['order_date'] = pd.to_datetime(df['order_date'])
df['start_date'] = pd.to_datetime(df['start_date'])
df['create_date'] = pd.to_datetime(df['create_date'])
```

```python
df.dtypes
```

```python
df['Month'] = df['order_date'].dt.month
```
