Scaling means transforming features so they are on a similar range.

Some algorithms perform poorly when fetures have very diffent scale.

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "EmployeeID":np.arange(1,101),
    "Age":np.random.randint(20,60,100),
    "Salary":np.random.randint(20000,100000,100)
   
}
df = pd.DataFrame(data)
```

```python
df.head()
```

<h2>Standard Scaler</h2>

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age_transformed','Salary_transformed']] = scaler.fit_transform(df[['Age','Salary']])
```

```python
df.head()
```

```python
avg = df['Salary'].mean()
std = df['Salary'].std()
```

```python
df['Salary_manual'] = (df['Salary'] - avg)/std
```

```python
df['Salary_Inverse_manual'] = (df['Salary_manual'] * std) + avg
```

```python
df.head()
```

```python
df[['age_inverse','salary_inverse']] = scaler.inverse_transform(df[['age_transformed','Salary_transformed']])
```

```python
df.head()
```

<h2>Min Max Scaler</h2>

```python
data = {
    "EmployeeID":np.arange(1,101),
    "Age":np.random.randint(20,60,100),
    "Salary":np.random.randint(20000,100000,100)
   
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Age_scaled','Salary_scaled']] = scaler.fit_transform(df[['Age','Salary']])
```

```python
df.head()
```

```python
xmin = df['Salary'].min()
xmax = df['Salary'].max()
```

```python
df['Salary_Transformed'] = (df['Salary'] - xmin)/(xmax-xmin)
```

```python
df['Salry_manual'] = df['Salary_Transformed']*(xmax-xmin)+xmin
```
