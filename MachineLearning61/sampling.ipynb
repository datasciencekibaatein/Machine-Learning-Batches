<h1>Sampling</h1>

Selecting a subset of data from a large population

```python
import numpy as np
import pandas as pd
```

```python

data = pd.DataFrame(
    {
        "customer_id":range(1,1001),
        "age":np.random.randint(18,65,1000),
        "income":np.random.randint(20000,200000, 1000),
        "purchased":np.random.choice([0,1],size=1000,p=[0.8,0.2])
    }
)
```

```python
data
```

```python
sample_random = data.sample(n=200,random_state=42)
```

```python
sample_random["purchased"].value_counts()
```

```python
sample_random 
```

```python
sample_random = data.sample(frac=0.2,random_state=34)
```

```python
sample_random["purchased"].value_counts()
```
