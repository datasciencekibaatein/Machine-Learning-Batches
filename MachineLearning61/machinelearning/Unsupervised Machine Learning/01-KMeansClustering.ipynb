<h1><strong>KMeans Clustering</strong></h1>

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X,_=make_blobs(n_samples=300,centers=4,random_state=42)
```

```python
kmeans = KMeans(n_clusters=4,random_state=42)
labels = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=_)
plt.show()
```

Predicted datapoints

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=20,marker='x',c='red')
plt.show()
```

<h3><strong>Choosing K(Elbow method)</strong></h3>

```python
w= []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X)
    w.append(kmeans.inertia_)

plt.plot(range(1,10),w)
plt.xlabel("K")
plt.ylabel("W")
plt.title("Elbow method")
plt.show()
```

<strong>silhouette Score</strong>

```python
from sklearn.metrics import silhouette_score

for k in range(2,10):
    kmeans = KMeans(n_clusters=k,random_state=42,max_iter=100)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X,labels)
    print(f"K={k}, Score={score}")
```
