<h1>EDA(Exploratory data analysis)</h1>

```python
import numpy as np
import pandas as pd
```

```python
# np.random.seed(42)

rows = 350

data = {
    "Customer_ID": range(1001, 1001 + rows),
    "Age": np.random.randint(18, 65, rows),
    "Gender": np.random.choice(["Male", "Female"], rows),
    "City": np.random.choice(["Delhi", "Mumbai", "Bangalore", "Pune", "Hyderabad"], rows),
    "Annual_Income": np.random.randint(200000, 2000000, rows),
    "Spending_Score": np.random.randint(1, 101, rows),
    "Product_Category": np.random.choice(["Electronics", "Clothing", "Grocery", "Furniture"], rows),
    "Purchase_Amount": np.random.randint(500, 50000, rows),
    "Payment_Method": np.random.choice(["Cash", "Card", "UPI"], rows),
    "Satisfaction": np.random.randint(1, 6, rows)
}
```

<h3>Convert Data into dataframe</h3>

```python
df = pd.DataFrame(data)
```

Histigram plot

```python
import matplotlib.pyplot as plt
```

Age distribution of the customers

```python
plt.hist(df["Age"],bins=15)
plt.xlabel("Age")
plt.ylabel("count")
plt.title("Age Distribution")
plt.show()
```

To chech the no. of male and female

```python
df["Gender"].value_counts()
```

Purchase amount Distribution

```python
plt.boxplot(df["Purchase_Amount"])
plt.title("Purches amount box plot")
plt.show()
```

```python
df.head()
```

```python
df.groupby(["Product_Category","Gender"])["Purchase_Amount"].sum()
```

```python
df.groupby(["Product_Category","Gender"])["Purchase_Amount"].agg(["sum","mean","min"])
```

```python
df.groupby(["Product_Category","Gender"]).agg({"Purchase_Amount":["mean","sum"],
                                               "Satisfaction":["min","max","mean"]}).reset_index()
```

```python
df.groupby(["Product_Category","Gender"]).agg(
    Avg_purchaseAmount = ("Purchase_Amount","mean"),
    Revenue = ("Purchase_Amount","sum"),
    AvgSatisfaction = ("Satisfaction","mean")
).reset_index()
```

```python
df.groupby("City")["Purchase_Amount"].sum().reset_index()
```
