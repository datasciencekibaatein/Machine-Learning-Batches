In simple linear regression we are having single independent variable/feature.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
df = pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
df.head()
```

```python
df['target'] = housing.target
```

```python
df.head()
```

```python
x = df[["MedInc"]] #Independent variable
```

```python
y = df["target"] #Dependent variable
```

Split data into training and testing part

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

Plot data points

```python
plt.scatter(x["MedInc"],y,color="red",alpha=0.5)
plt.xlabel("Income")
plt.ylabel("Price")
plt.show()
```

Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Train the linear model

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

model = reg.fit(x_train,y_train)
```

Predict the value

```python
pred_y = model.predict(x_test)
```

```python
pred_y
```

```python
plt.scatter(y_test,pred_y)
plt.xlabel("y test data")
plt.ylabel("predicte data")
```

Draw the best fit line

```python
plt.plot(x_test,pred_y,color="red")
plt.scatter(x_train,y_train)
plt.xlabel("income")
plt.ylabel("house price")
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred_y))
```
