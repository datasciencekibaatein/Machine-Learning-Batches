<center><h1 style="color:cyan;"><strong>Logistic Regression </strong></h1></center>

Logistic regression is a supervised machine learning algorithm used for classification problems.

Unlike linear regression which predicts continuous values, logistic regression predicts probablities of classes.

Example: 

|Problem | Target|
|------|-----|
|Email Spam detection | Spam/ Not Spam|
|Disease prediction | Sick / Healthy |
|Customer Churn | Leave / Stay |
|Fraud Detection | Fraud / Not Fraud|


So logistic regression answers.

<strong>What is the probablity that this data belongs to class 1 ?</strong>

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
data = load_breast_cancer()

df = pd.DataFrame(data.data,columns=data.feature_names)
df['target'] = data.target

print(df.head(10))
print(df.tail(10))
```

<h2><strong>Why do we need logistic regression:</strong></h1>

If we try to use linear regression for all classification, problem occur:

Example:

Suppose we predict if a student passes (1) or fails(0).

Linear regression might predict:
- 1.2
- -0.3

But probablities must lie between:

0<=P<=1

<strong style="color:magenta;">Linear regression can not guarantee that.</strong>

Therefore we use a function that compresses values between 0 and 1.

This is where the Sigmoid function comes in.
