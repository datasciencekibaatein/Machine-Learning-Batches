<center><h1 style="color:cyan;"><strong>Ridge Regression<code>(L2 Regularization)</code></strong></h1></center>

Ridge regression is a regularized version of Linear Regression.

It adds a penalty term to the cost function to reduce overfitting.

Large Lamda -> Coefficients shrink more <br/>
Small lamda -> Behave like linear regression

<strong>Why Ridge regression is used</strong>

- Multicollinearity Exists
- Model is overfitting
- Too many features
- Coefficients are unstable

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
data = load_diabetes()
x = data.data
y = data.target

print(x.shape)
print(y.shape)
```

Train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

Feature scaling

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Train ridge model

```python
regression = Ridge(alpha=1.0) #lambda = 1
model = regression.fit(x_train,y_train)
```

Predict values

```python
y_pred = model.predict(x_test)
```

R2 Score

```python
score = r2_score(y_test,y_pred)
print(score)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(x_test[:,0],y_test)
plt.scatter(x_test[:,0],y_pred,color='red')
plt.xlabel("x points")
plt.ylabel("y points Red-Predict, Blue-Actual")
plt.show()
```
