<h1><strong>Decision Tree</strong></h1>

Decision Tree is a supervised Machine learning algorithm.

Used for:

- Classification
- Regression

<h3><strong>Key Terminonlogies</strong></h3>

- Root Node -> Starting Point
- Decision Node -> Condition (eg, Age>30)
- Leaf Node -> Final Output
- Branch -> Outcome of condition

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn. model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_iris()
x = data.data
y= data.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=3,
    random_state=42
)
```

```python
model = model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
data.target_names
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8))
plot_tree(model,
          feature_names=data.feature_names,
          filled=True,
          class_names=data.target_names)

plt.show()
```
