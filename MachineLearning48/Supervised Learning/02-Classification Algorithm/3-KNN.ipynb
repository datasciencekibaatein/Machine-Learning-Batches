How KNN Works?

- Choose K (number of neighbors)
- Calculate distance between new point and all training point
- Pick th K nearest point
- Take majority vote (classification) or average (regression)

Small K -> Overfitting<br>
Large K -> Underfitting

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
data.feature_names
```

```python
X[0:20,]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
plt.show()
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test)
plt.show()
```

```python
accuracy
```

```python
ns = [2,3,4,5,6,7,8,9,10]
a_s = []


for n in ns:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    a_s.append(accuracy_score(y_test,y_pred))
```

```python
a_s
```

Advantages

- Simple & easy to understand
- No training phase (lazy)
- Works well with small datasets

Disadvantages

- Slow for large datasets
- Sensitive to noise
- Needs feature scaling
- Memory intensive
