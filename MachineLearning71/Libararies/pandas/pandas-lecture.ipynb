<h2><strong>concat</strong></h2>

```python
df1 = pd.DataFrame({
    "A":[1,2,3],
    "B":[4,5,6],
    "C":[7,8,9]
},index=[1,2,3]
)

df2 = pd.DataFrame({
    "D":[1,2,3],
    "B":[7,6,5],
    "C":[4,3,2]
},index=[4,5,6]
)
```

```python
arr = [df1,df2]

pd.concat(arr,axis=0)
```

<h2><strong>Join</strong></h2>

```python
pd.concat(arr,axis=0,join='outer')
```

```python
fact = pd.read_csv("fact_sales.csv")
customers = pd.read_csv("dim_customers.csv")
products = pd.read_csv("dim_products.csv")
```

```python
data = pd.merge(fact,customers,how="inner",on="customer_key")[["order_number","product_key","order_date","shipping_date","sales_amount","quantity","customer_id","first_name","country","marital_status","gender"]]
```

```python
fact_data = pd.merge(data,products,how="inner",on="product_key")[
    ["order_number","order_date","shipping_date","sales_amount","quantity","customer_id","first_name","country","gender","marital_status","product_number","product_name","category","subcategory","product_line"]]
```

```python
data
```

```python
fact_data
```

```python
fact_data.info()
```

```python
fact_data["order_date"] = pd.to_datetime(fact_data["order_date"])
```

```python
fact_data
```

```python
fact_data["shipping_date"] = pd.to_datetime(fact_data["shipping_date"])
```

```python
fact_data["year"] = fact_data["order_date"].dt.year
fact_data["month"] = fact_data["order_date"].dt.month
```

```python
fact_data.groupby(["year","month_name"])["sales_amount"].sum()
```

```python
fact_data["sales_amount"].sum()
```

```python
month_map = {
    1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",7:"jul",8:"aug",9:"sep",10:"oct",11:"nov",12:"dec"
}
```

```python
fact_data["month_name"] = fact_data["month"].map(month_map)
```

```python
f_data = fact_data[
    (fact_data["sales_amount"]==fact_data["sales_amount"].min())
                            &    
    (fact_data["sales_amount"].notna())]
```
