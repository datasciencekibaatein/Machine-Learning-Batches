```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
```

<h2>Distribution plot</h2>

Histogram

```python
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.show()
```

KDE

```python
sns.kdeplot(df['total_bill'],fill=True)
plt.show()
```

displot

```python
sns.displot(df['total_bill'],kde=True)
plt.show()
```

displot and histogram both are similar

<h2>Categorical Plot</h2>

Bar plot

```python
sns.barplot(x='day',y='total_bill',data=df)
plt.show()
```

countplot

```python
sns.set_theme(style='darkgrid',palette='viridis')
sns.countplot(x='day',data=df)
plt.show()
```

boxplot

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=df)
plt.show()
```

Strip plot

```python
sns.stripplot(x='day',y='total_bill',data=df)
plt.show()
```

Swarm plot

```python
sns.swarmplot(x='day',y='total_bill',data=df)
plt.show()
```

<h2>Relation Ship plots</h2>

Scatter plot

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='day')
plt.show()
```

Line plot

```python
sns.lineplot(x='size',y='total_bill',data=df)
plt.show()
```

Regplot

```python
sns.regplot(x='total_bill',y='tip',data=df)
plt.show()
```

<h2>Matrix Plot</h2>

heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.show()
```

Cluster Map

```python
sns.clustermap(corr)
```

<h2>Multi Plot</h2>

```python
sns.pairplot(df,hue='day')
```

Facet Plot

```python
g = sns.FacetGrid(df,col='smoker')
g.map(sns.histplot,"total_bill")
```

joint plot

```python
sns.set_theme(style='darkgrid',palette='viridis')
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
plt.show()
```

rug plot

```python
sns.rugplot(df['total_bill'])
sns.kdeplot(df['total_bill'])
plt.show()
```

```python
plt.style.use("dark_background")

plt.scatter(x=df['tip'],y=df['total_bill'])
sns.lineplot(x='tip',y='total_bill',data=df)
plt.title("Scatter Plot",color='teal')
plt.show()
```
