<h1><strong>What are missing values?</strong></h1>

We ecounter cells that are empty, null or undefined. These are called missing values - data point that were expected but were not recorded.

In python and pandas, missing values appear as NaN (Not a Number), None, or somtime as string like "N/A", "-" or "unknown" that must be manually converted .

<h2><strong>What missing values occurs </strong></h2>

- Survey non-response
- System failure
- Data entry errors
- Not applicable
- Deliberate omission
- Data migration


<h2><strong>Types of missing values</strong></h2>

- MCAR - Missing completely at random
- MAR - Missing at random
- MNAR - Missing not at random

<h2><strong>1. MCAR(Missing completely at random)</strong></h2>

The probability of a value being missed has absolutely no relationship with any variable in the dataset.

<h2><strong>2. MAR(Missing at Random)</strong></h2>

The missingness is related to observed data, but not the missing values themselves.

<h2><strong>3. MNAR(Missing not at random)</strong></h2>

the missingness depends on the observed value itself. This is hardest to handle.

    2.1 Univariate Imputation<br>

```python
import pandas as pd
import numpy as np

data = {
    'Name':["Rohan","Krish","Bob","Sohit","Kashish","Mayank"],
    'Age':[24,np.nan,34,23,np.nan,40],
    'Score':[85,90,np.nan,88,92,95],
    'Attendance':[100,90,80,np.nan,90,85]
}



df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull().sum()
```

<h2><strong> Rows Deletion</strong></h2>

<h3><strong>When it is good to use?</strong></h3>

- Low percentage: If < 5% data is missing
- MCAR: The missingness has no hidden pattern.
- Large Dataset: If you have 1 million rows, losing 1000 wont hurt your model's perfomance.

<h3><strong>When not to use it?</strong></h3>

- Small Dataset: If you only have 100 rows, every single row is precious
- Bias Risk: If only "High income" individuals refused to answer the survey, dropping those rows removes your "High income" representation entirely.

<h2><strong>Column Deletion</strong></h2>



<h3><strong>When it is good to Use</strong></h3>

- High missingness: eg. > 50, 60 ,....
- Irrelevance: If the feature isn't crucial for the prediction

<h3><strong>When it is not recommended</strong></h3>

- Domain Importance

```python
df['Notes'] = [np.nan,"Good",np.nan,np.nan,np.nan,"Agerage"]
```

```python
df
```

```python
missing_pct = df.isnull().sum()/len(df) * 100
```

```python
missing_pct.values
```

```python
limit = 30

cols_to_drop = missing_pct[missing_pct>limit].index

df_reduced = df.drop(columns=cols_to_drop)
```
