<h1><strong>Skewness</strong></h1>

Skewness measures the asymmetry of your data's distribution. Instead of perfectly balanced "bel curve", the data leans heavily to one side, stretching out into a long tail.

<h2><strong>Types of skewness</strong></h2>

- Zero (symmetric):Perfectly balanced. Mean = Median
- Positive (Right Skewed): Tail stretches to the right. Mean > Median
- Negative (Left Skewed) : Tail stretches to the left. Mean < Median

```python
import numpy as np
import matplotlib.pyplot as plt
```

<strong>Symmetric Distribution</strong>

```python
np.random.seed(42)

symmetric_data = np.random.normal(loc=0,scale=1,size=2000)

plt.figure(figsize=(6,4))
plt.hist(symmetric_data,bins=30,)
plt.title("Symmetric distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
np.mean(symmetric_data)
```

```python
np.median(symmetric_data)
```

<strong>Left Skewed</strong>

```python
np.random.seed(42)
left_skewed = -np.random.gamma(shape=2.0,scale=1.5,size=2000)
plt.figure(figsize=(6,4))
plt.hist(left_skewed,bins=30)
plt.title("Left skewed distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
np.mean(left_skewed)
```

```python
np.median(left_skewed)
```

<strong>Right Skewed</strong>

```python
np.random.seed(42)
right_skewed = np.random.gamma(shape=2.0,scale=1.5,size=2000)
plt.figure(figsize=(6,4))
plt.hist(right_skewed,bins=30)
plt.title("Rightskewed distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
np.mean(right_skewed)
```

```python
np.median(right_skewed)
```

<h2><Strong>Right skewed</strong></h2>

Positive skewness occurs when a distribution has a long, extended tail on the right side of the x-axis, the majority of the data points are clustered at the lower left end.

```python
import pandas as pd

np.random.seed(42)

salaries = np.random.exponential(scale=50000,size=1000)

df_pos = pd.DataFrame({"Salary":salaries})
```

```python
df_pos
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df_pos['Salary'],kde=True,color='magenta')
mean_val = df_pos['Salary'].mean()
median_val = df_pos['Salary'].median()

plt.axvline(mean_val,color='red',linestyle='--',label=f'Mean:{mean_val:.0f}')
plt.axvline(median_val,color='green',linestyle='-',label=f'Median:{median_val:.0f}')

plt.title("Positve skewed")
plt.legend()
plt.show()
```

```python
df_pos['log_Salary'] = np.log1p(df_pos['Salary'])
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df_pos['log_Salary'],kde=True,color='magenta')
mean_val = df_pos['log_Salary'].mean()
median_val = df_pos['log_Salary'].median()

plt.axvline(mean_val,color='red',linestyle='--',label=f'Mean:{mean_val:.0f}')
plt.axvline(median_val,color='green',linestyle='-',label=f'Median:{median_val:.0f}')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Positve skewed: Fixed")
plt.legend()
plt.show()
```

<h2><strong>Left skewed (Negative skewness)</strong></h2>

Negative skewness features a long tail stretching out to the left side. Here ,the vast majorty of the data is clustered at the higher (right) end.

```python
scores = 100- np.random.exponential(scale=15,size=1000)
df_neg = pd.DataFrame({'Score':scores})
df_neg.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_neg['Score'],kde=True,color='salmon')

mean_val = df_neg['Score'].mean()
median_val= df_neg['Score'].median()

plt.axvline(mean_val,color='red',linestyle='--',label=f'Mean:{mean_val:.0f}')
plt.axvline(median_val,color='green',linestyle='-',label=f'Median:{median_val:.0f}')
plt.legend()
plt.title("Negative skewed")
plt.show()
```

```python
df_neg['Squared_Score'] = np.square(df_neg['Score'])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_neg['Squared_Score'],kde=True,color='teal')

mean_val = df_neg['Squared_Score'].mean()
median_val= df_neg['Squared_Score'].median()

plt.axvline(mean_val,color='red',linestyle='--',label=f'Mean:{mean_val:.0f}')
plt.axvline(median_val,color='green',linestyle='-',label=f'Median:{median_val:.0f}')
plt.legend()
plt.title("Negative skewed : Fixed")
plt.show()
```

```python
df_neg['Score'].skew()
```

<h2><strong>Numerical Detection<strong></h2>

- -0.5 and 0.5 : good to go
- -1 and -0.5 / 0.5 : Moderately skewed
- -1 or greater than 1.0 highly skewed (requries transforms)
