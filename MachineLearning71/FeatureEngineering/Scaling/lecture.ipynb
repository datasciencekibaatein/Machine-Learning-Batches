Transforming features so they are on similar scale.

<h1><strong>Standard Scaler</strong></h1>

```python
import pandas as pd
import numpy as np

data = {
    'age':np.random.randint(20,55,100),
    'salary':np.random.randint(30000,100000,100)
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
pip install scikit-learn
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[["age_scaled","salary_scaled"]] = scaler.fit_transform(df[['age','salary']])
```

```python
df.head()
```

<code>Z-score Formula:</code>

$$Z =\frac{X - \mu}{\sigma} $$

```python
df['age_scaled_manual'] = (df['age'] - df['age'].mean())/df['age'].std()
```

```python
df['age_inverse_manual'] = df['age_scaled_manual']*df['age'].std() + df['age'].mean() 
```

```python
df.head()
```

```python
df[['age_inverse','salary_inverse']] = scaler.inverse_transform(df[['age_scaled','salary_scaled']])
```

```python
df.head()
```

<h1><strong>Min-Max Scaler</strong></h1>

<code>Min-Max Scaler Formula: </code>
$$X_{scaled} = \frac{X - X_{min}}{X_{max}-X_{min}}$$

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['age_mimx','salary_mimx']] = scaler.fit_transform(df[['age','salary']])
```

```python
df.head()
```

```python
minn = df['salary'].min()
maxx = df['salary'].max()
```

```python
df['salary_mimx_manual'] =( df['salary'] - minn )/(maxx - minn)
```

```python
df.head()
```

```python
df['salary_inverse_mimx'] = df['salary_mimx_manual'] * (maxx-minn) + minn
```

```python
df.head()
```

```python
df[['age_inverse_scaler','salary_inverse_scaler']] = scaler.inverse_transform(df[['age_mimx','salary_mimx']])
```
