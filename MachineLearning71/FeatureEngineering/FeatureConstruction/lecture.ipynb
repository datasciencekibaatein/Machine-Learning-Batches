```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 100

df = pd.DataFrame(
    {
        "age":np.random.randint(18,55,n),
        "salary":np.random.randint(30000,100000,n),
        "experience":np.random.randint(1,30,n),
        "hours_worked":np.random.randint(4,12,n)
    }
)
```

```python
df.head()
```

```python
df['salary_per_experience'] = df['salary']/df['experience']
```

```python
df['age_experience_interaction'] = df['age']*df['experience']
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins = [18,28,42,60],
    labels=['Young','Middle','Senior']
)
```

```python
df['log_salary'] = np.log(df['salary'])
```
