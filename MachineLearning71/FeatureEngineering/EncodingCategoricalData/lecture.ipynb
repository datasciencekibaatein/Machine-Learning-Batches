<h1 style="color:magenta;"><strong>Encoding Categorical data</strong></h1>

Encoding converts categorical/text data into numerical features/format.

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 100

df= pd.DataFrame({
    "age":np.random.randint(18,60,n),
    "salary":np.random.randint(30000,100000,n),
    "city":np.random.choice(["Delhi","Mumbai","Pune","Bareilly"],n),
    "gender":np.random.choice(["Male","Female"],n),
    "education":np.random.choice(['High School',"Graduate","Post Graduate"],n)
})

df.head()
```

<h2>Label Encoder</h2>
Label encode assign unique number to each category

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['City_label'] = le.fit_transform(df['city'])
```

<h2>One hot encoder</h2>

Give you sparese matrix(0,1) and generate spearate columns to each category

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder()

encoded= oe.fit_transform(df[['gender']]).toarray()
```

```python
df[['Female','male']] = encoded
```

<h2>Ordinal Encode: Used when categories have order</h2>

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(
    categories=[[
        "High School",
        "Graduate",
        "Post Graduate"
    ]]
)

df['education_encoded'] = oe.fit_transform(df[['education']])

df.head(10)
```
