<h2><strong>Operators</strong></h2>

Operators are the special symbols we use to perform computation on variables and values.

- Arithmetic operators ( + , - , * , / , % , ** , //)
- Comparision Operators (==, != , > , < , <= , >= ) , comparision operators returns Boolean values
- Logical Operators (and ,or , not)
- Assignment Operators (= , += , -=, *= , /= , %=, **=, //= , &= , |=, ~=, <<=, >>=)
- Identity Operators (is , is not)
- Membership Operators (in , not in)
- Bitwise operators (&, | , ^ , ~, << , >>)

<h2>Arithmetic operators</h2>

```python
num1 = 66
num2 = 49

print("Addtion: ",num1+num2)#addition
print("Subtraction: ",num1 - num2)#subtraction
print("multiplication: ",num1*num2)#multiplication
print("Division: ",num1/num2)#division
print("Modulus: ",num1%7)#remainder
print("Power: ",6**3)#exponentiation(power)
print("Floor division: ",num1//num2)
```

<h2>Logical Operators</h2>

```python
age = 21
salary = 35000
review = "Excellent"
#return true/false if age is greater than 20 and salary is greater than 40000
print(age > 20 and salary > 40000 and review == "Excellent") # if all the condition are true then only return true otherwise false if any of fails
```

```python
#or
# if any of condition become true among given, it will return u True, if all of them are false, it returns false
print(age > 20 or salary > 40000 or review == "Excellent")
```

```python
#not
# if some condition returns True it will make it false, if condition is false it will make it True
print(not age>20)
```

<h2>Bitwise operators</h2>

```python
num1 = 10 #1010
num2 = 11 #1011

#&
print(num1 & num2)

#|
print(num1 | num2)

#^
print(num1 ^ num2)

#~ (~n = -(n+1))
print(~num2)


#<<
print(num2<<2)

#>>
print(num2>>4)
```

<h2>Identity Operators</h2>

```python
#is


lst = [2,3,1]
lst2 = [2,3,1]

temp = lst
print(lst is lst2)
print(lst is temp)
```

```python
print(lst is not lst2)
print(lst is not temp)
```
