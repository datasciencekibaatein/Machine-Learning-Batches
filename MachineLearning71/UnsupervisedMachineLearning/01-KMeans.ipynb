## KMeans Clustering

K-Means is an unsupervised ml algorithm that groups similar data points into K Clusters.

### The algorithms tries to :

- Minimize distance within a cluster.
- Maximize distance between cluster.

### The word:

- K = Number of clusters
- Means = Cluster center (average value)

### Algorithms:

- Step01: Initialize K Centroids randomly.
- Step02: Assign every point to nearest centroid.
- Step03: Update the centeroid position
- Step04: Repeat steps 2 and 3 unitl convergence.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris


df = pd.DataFrame(load_iris().data,columns=load_iris().feature_names)

df.head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    c="teal"
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
k = 3
```

```python
kmeans = KMeans(n_clusters=k,random_state=42)

kmeans.fit(df)
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    c=load_iris().target,
    cmap='viridis'
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Actual clusters")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    c=labels,
    cmap='viridis'
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show("Predicted clusters")

plt.show()
```

```python
centroids = kmeans.cluster_centers_

print(centroids)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    c=labels,
    cmap='viridis'
)

plt.scatter(
    centroids[:,0],
    centroids[:,1],
    s=300,
    c = 'red',
    marker='x'
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show("Predicted clusters")

plt.show()
```

### WCSS Method(Within-cluster sum of squares) in KMeans

WCSS is the sum of squared distances between each data points and the centroids of its assigned

```python
kmeans.inertia_
```

```python
inertia = []

for k in range(1,11):
    model = KMeans(
        n_clusters=k,
        random_state=42
    )
    
    model.fit(df)
    inertia.append(model.inertia_)
```

```python
inertia
```

```python
plt.figure(figsize=(8,6))
plt.plot(
    range(1,11),
    inertia,
    marker='o'
)

plt.xlabel("Number of clusters(K)")
plt.ylabel("WCSS")

plt.title("Elbow method")

plt.show()
```
