# PCA(Principal component analysis)

PCA is for unsupervised learning

PCA transforms original features into new features called Principal Components (PCs)


These components:

- Are linear combinations of orginal fetures
- Capture maximum variance

PCA finds PC1 and PC2

- PC1: Direction of maximum spread
- PC2: Direction of perpendicular to PC1

### Goal of PCA

Find a direction wehre variance is maximum.

Higher variance => More Information.

Low variance dimensions usually contains:

- Noise
- Redundant information

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
import matplotlib.pyplot as plt

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c = y
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
