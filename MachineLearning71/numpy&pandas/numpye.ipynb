Random shuffle

```python
arr = np.array([1,2,3,4,5,6])
np.random.shuffle(arr)
print(arr)
```

Random permutation

<code>Similar to shuffle but does not nodify original array.
Original array remains unchanged</code>

```python
arr = np.array([1,2,3,4,5,6])
print(np.random.permutation(arr))

print(arr)
```

Setting random seed

```python
np.random.seed(42)

np.random.rand(3)
```
