```python
import numpy as np
```

```python
#1D array

arr = np.array([23,45,67,23])
print(type(arr))
```

```python
lst = [45,34,67,90,23]
arr = np.array(lst)
print(type(arr))
```

```python
#2D array

mat = [[2,3,4],
       [4,5,6],
       [7,8,9]]

matrix = np.array(mat)
print(matrix)
print(type(matrix))
print(matrix.ndim)
print(arr.ndim)
```

```python
values = [('Arthur', 1.8, 41), 
          ('Lancelot', 1.9, 38),
          ('Galahad', 1.7, 38)]
```

```python
dtype = [('name', 'S10'), ('height', float), ('age', int)]
```

```python
data= np.array(values,dtype= dtype)
print(data)
```

```python
np.sort(data,order='name')
```
